```python
import pandas as pd

df = pd.read_csv("./results.csv")
df.head()
df = df[df["Leaf Size"] != 500]
df.columns = ['Airport', 'Leaf Size', 'Linear', 'Proposed', 'Lund']
```

```python
%matplotlib inline

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)



np.random.seed(sum(map(ord, "distributions")))
airports = {"eddt": "Berlin Tegel Airport", 
            "egll": "Heathrow Airport", 
            "yssy": "Sydney Airport", 
            "zbaa": "Beijing Capital International Airport", 
            "lebl": "BarcelonaEl Prat Airport", 
            "lfpg": "Charles de Gaulle Airport", 
            "limc": "Milan-Malpensa Airport"}

ylims = [(3.7,4.1), (3, 3.35), (2.55, 2.8), (3.2, 4), (2.1, 2.6), (3.5, 3.9), (3.7, 3.95)]
for i, code in enumerate(["eddt", "egll", "yssy", "zbaa", "lebl", "lfpg", "limc"]): 
    df.loc[df['Airport'] == code].plot(title=airports[code], legend=False, kind="bar", x="Leaf Size", ylim=ylims[i], fontsize=16)
```
